```python
#importing required packages
import requests
import bs4
```

```python
res = requests.get('https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_1?ie=UTF8&pg=1')
```

```python
content = res.content
soup = bs4.BeautifulSoup(content)
```

```python
complete_data = []
for i in soup.findAll('div',attrs={'class':'a-section a-spacing-none aok-relative'}):
    #print(i.text)
    book_name = i.find('span',attrs={'class':'zg-text-center-align'})
    
    name = book_name.find_all('img',alt=True) # Fetching Book name
    author = i.find('a', attrs={'class':'a-size-small a-link-child'}) #author name
    rating = i.find('span', attrs={'class':'a-icon-alt'}) #rating
    no_ratings = i.find('a', attrs={'class':'a-size-small a-link-normal'}) #no of ratings
    price = i.find('span', attrs={'class':'p13n-sc-price'}) 
    
    myList = []
    
    #Name of the Book
    if name is not None:
        myList.append(name[0]['alt'])
    else:
        myList.append('Unknown Book Name')
        
    #Author name
    if author is not None:
        myList.append(author.text)
    else:
        myList.append('Unknown Author')
        
    
    #Book rating
    if rating is not None:
        myList.append(rating.text)
    else:
        myList.append('No Rating')
        
        
    #No of ratings
    if no_ratings is not None:
        myList.append(no_ratings.text)
    else:
        myList.append('No Rating')
        
        
    #No of ratings
    if price is not None:
        myList.append(price.text)
    else:
        myList.append('Price not mentioned')
        
    complete_data.append(myList)
```
